```python
# importing modules
import pandas as pd
import editdistance
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# It is observed that same instances may look different just due to some human errors or conventional naming formats.
# like small change due to the presence or absence of last name suffixes Jr, III(different naming format) or same instances
# may look different due to one or two characters change in their name(typing error).It is also observed that people with more
# than one word in their first and last name often either skip to write second word or end up writing first character of second 
# word. Taking all these observed errors into account function named match is created that output 0 if the observed variation 
# is due to human error or conventional naming otherwise output 1.
 
def match(fn1,fn2,ln1,ln2):
      # small relaxation for not meeting the exact first and last name with edit distance less than 2 for both first and last name
    if (editdistance.eval(fn1.split(" ")[0],fn2.split(" ")[0])<=2)&(editdistance.eval(ln1.split(" ")[0],ln2.split(" ")[0])<=2):
        return 0
    else:
        return 1
# function that find the edit distance per unit average length of the full name.

def ratio(x,y,z):
    return 2*z/(len(x)+len(y))
```

```python
# loading supervised data-set into df1
df1=pd.read_csv("sample_input.csv")
df1
```

```python
df1.shape
```

```python
# feature engineering using edit distance as one of the metric

df1['name_dist'] = list(map(editdistance.eval,df1.fullname1,df1.fullname2))
df1['dob_dist'] = list(map(editdistance.eval,df1.dob1,df1.dob2))
df1['gn_dist'] = list(map(editdistance.eval,df1.gn1,df1.gn2))
# it was observed that people with long names often skip to write their full name and end up writing only first few characters
# of the second word of the first and last name
df1["feat"]=list(map(match,df1.fn1,df1.fn2,df1.ln1,df1.ln2))
# edit distance between the names for the average unit length of the name or in other words edit distance per unit length of the full name.
df1["ratio"]=list(map(ratio,df1.fullname1,df1.fullname2,df1.name_dist))
df1
```

```python
# data-set for training (I have omitted the name_dist feature) beacuse ratio feature is calculated from it. 

df2=df1[["response","dob_dist","gn_dist","feat","ratio"]]
```

```python
df2
```

```python
# seperation into feature matrix and responce variable

X=df2.drop("response",axis=1)

# instances are labelled such that 0 for being same and 1 for being different
Y=df2.response
```

```python
# assuming that date of birth data is 100% accurate, so any variation(small or large DoB edit distance) should contribute equally. 
# this same operation is not performed on gender edit-distance because the variation are either 0 or 1. so our features are in
# the same range which will help our model to converge faster.

X["dob_dist"]=X["dob_dist"].map(lambda T: 1 if T>0 else 0)     

# training of model
lr=LogisticRegression()
lr.fit(X,Y)

pred=lr.predict(X)
scores=metrics.accuracy_score(Y,pred)
metrics.confusion_matrix(Y,pred)
```

```python
# loading origional data-set into df

df=pd.read_csv("origional_data_set.csv")
df["Full_Name"]=df["fn"]+" "+df["ln"]
df.head()
```

```python
# Identification of unique patients using logistic regression
# copy of the origional data-set is made to remove the duplicates of the instances on each iteration
reduced=df.copy()
# for-loop will run on all the instances of the origional data-set
for i in range(len(df)):
    s=df.copy()
    # pair-wise comparison of the I-th element of the data-set(df) with rest of the data-set.
    s["Name1"]=df.loc[i,"Full_Name"]
    s["dob1"]=df.loc[i,"dob"]
    s["fn1"]=df.loc[i,"fn"]
    s["ln1"]=df.loc[i,"ln"]
    s["gn1"]=df.loc[i,"gn"]
    # feature engineering 
    s['name_dist'] = list(map(editdistance.eval,s.Name1,s.Full_Name))
    s['dob_dist'] = list(map(editdistance.eval,s.dob1,s.dob))
    s['gn_dist'] = list(map(editdistance.eval,s.gn1,s.gn))
    s["feat"]=list(map(match,s.fn1,s.fn,s.ln1,s.ln))
    s["ratio"]=list(map(ratio,s.Name1,s.Full_Name,s.name_dist))
    d=s[["dob_dist","gn_dist","feat","ratio"]]
    # assuming that date of birth data is 100% accurate so any variation(small or large DoB edit distance) should contribute equally
    
    d["dob_dist"]=d["dob_dist"].map(lambda T: 1 if T>0 else 0)     
    s["prediction"]=lr.predict(d)
    # indexes of the instances that are predicted similar to the I-th element of the data-set(df)
    l=s[s.prediction==0.0].index
    l=list(l.values)
    # only remove the elements which are below the I-th element index or in other words larger than I-th element index
    l=list(filter(lambda x: x>i, l))

    try:
        reduced.drop(l,axis=0,inplace=True)
    except:
        pass
reduced
```
